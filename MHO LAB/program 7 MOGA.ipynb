# Program 7: Multi-Objective Optimization in Genetic Algorithm (MOGA)

**Objective:** Use MOGA to optimize multiple objectives such as maximizing energy output, minimizing construction cost, and reducing environmental impact using the DEAP library.

```python
import random
import numpy as np
from deap import base, creator, tools, algorithms
import matplotlib.pyplot as plt
```

```python
# -----------------------------
# 1. Define multi-objective fitness
# Maximize energy (+1), Minimize cost (-1), Minimize impact (-1)
# -----------------------------
creator.create("FitnessMulti", base.Fitness, weights=(1.0, -1.0, -1.0))
creator.create("Individual", list, fitness=creator.FitnessMulti)

toolbox = base.Toolbox()
```

```python
# -----------------------------
# 2. Decision variables
# -----------------------------
toolbox.register("attr_float", random.uniform, 0, 10)
toolbox.register("individual", tools.initRepeat, creator.Individual,
                 toolbox.attr_float, n=3)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
```

```python
# -----------------------------
# 3. Objective functions
# -----------------------------
def evaluate(individual):
    size, material, tech = individual

    # Toy objective models
    energy = 2 * size + 3 * tech
    cost = 5 * size + 2 * material
    impact = size + material + tech

    return energy, cost, impact

toolbox.register("evaluate", evaluate)
```

```python
 #-----------------------------
# 4. Genetic operators
# -----------------------------
toolbox.register("mate", tools.cxBlend, alpha=0.5)
toolbox.register("mutate", tools.mutGaussian, mu=0, sigma=1, indpb=0.2)
toolbox.register("select", tools.selNSGA2)
```

```python
# -----------------------------
# 5. Run MOGA
# -----------------------------
population = toolbox.population(n=50)

algorithms.eaMuPlusLambda(
    population,
    toolbox,
    mu=50,
    lambda_=100,
    cxpb=0.7,
    mutpb=0.3,
    ngen=40,
    verbose=True
)
```

```python

# -----------------------------
# 6. Extract Pareto-optimal solutions
# -----------------------------
pareto_front = tools.sortNondominated(population, k=len(population), first_front_only=True)[0]

print("Pareto-optimal solutions:")
for ind in pareto_front[:5]:
    print(ind, ind.fitness.values)
```
