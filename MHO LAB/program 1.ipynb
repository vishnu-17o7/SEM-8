```python
import math
import matplotlib.pyplot as plt
```

```python
xmin = -50
xmax = 50
graph_x = []
graph_y = []
```

```python
for x in range(xmin, xmax + 1):
    obj_value = x**2 - 4*x + 5 #+ math.sin(x**2/2) + 0.5*math.cos(2*x)
    graph_x.append((x))
    graph_y.append(obj_value)
    print(f"x: {x}, Objective Function Value: {obj_value}")
```

```python
# print(values)
plt.plot(graph_x, graph_y)
plt.show()
```

```python
#Global Minimum
min_value = min(zip(graph_x, graph_y), key=lambda item: item[1])
print(f"Global Minimum occurs at x = {min_value[0]} with value = {min_value[1]}")
```

```python
pi = math.pi

x_list = [-pi/2, -pi/4, 0, pi/4, pi/2]
graph_x = []
graph_y = []
```

```python
for x in x_list:
    obj_value = math.sin(x**2/2) + 0.5*math.cos(2*x)
    graph_x.append((x))
    graph_y.append(obj_value)
    print(f"x: {x}, Objective Function Value: {obj_value}")
```

```python
plt.plot(graph_x, graph_y)
plt.show()
```

```python
#Global Minimum
min_value = min(zip(graph_x, graph_y), key=lambda item: item[1])
print(f"Global Minimum occurs at x = {min_value[0]} with value = {min_value[1]}")
```
