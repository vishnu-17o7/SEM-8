```python
import nltk
from nltk.corpus import treebank, stopwords
from nltk.collocations import BigramCollocationFinder, TrigramCollocationFinder
from nltk.metrics import BigramAssocMeasures, TrigramAssocMeasures

# Download necessary datasets
nltk.download('treebank')
nltk.download('stopwords')
nltk.download('punkt')
```

```python
# Load words from the WSJ (Treebank) corpus
wsj_words = treebank.words()

# 1. Initialize the Bigram Finder
bigram_finder = BigramCollocationFinder.from_words(wsj_words)

# 2. Apply Filters (Remove stopwords and short words/punctuation)
ignored_words = set(stopwords.words('english'))
bigram_finder.apply_word_filter(lambda w: len(w) < 3 or w.lower() in ignored_words)

# 3. Apply Frequency Filter (Only consider bigrams that appear at least 5 times)
bigram_finder.apply_freq_filter(5)

# 4. Score Bigrams using Likelihood Ratio (identifies strong associations)
bigram_measures = BigramAssocMeasures()
top_bigrams = bigram_finder.nbest(bigram_measures.likelihood_ratio, 10)

print("Top 10 Bigram Collocations in WSJ:")
for bg in top_bigrams:
    print(f"- {' '.join(bg)}")
```

```python
# 1. Initialize the Trigram Finder
trigram_finder = TrigramCollocationFinder.from_words(wsj_words)

# 2. Apply Filters
trigram_finder.apply_word_filter(lambda w: len(w) < 3 or w.lower() in ignored_words)
trigram_finder.apply_freq_filter(3)

# 3. Score Trigrams
trigram_measures = TrigramAssocMeasures()
top_trigrams = trigram_finder.nbest(trigram_measures.likelihood_ratio, 10)

print("\nTop 10 Trigram Collocations in WSJ:")
for tg in top_trigrams:
    print(f"- {' '.join(tg)}")
```
