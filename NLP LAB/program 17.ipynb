## Finding Similar Vectors Using t-SNE with GloVe and Visualizing
Find similar vectors on t-SNE class with GloVe and visualize it with plots.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import urllib.request
import zipfile
import os

# Download a small version of GloVe if not present (for demonstration we use a subset)
# Note: Full GloVe is large, so we will simulate GloVe embeddings for a small vocabulary
# to keep the demonstration fast, or you can point to a local glove.6B.50d.txt file.

print("Generating sample embeddings to represent GloVe vectors...")
words = ['king', 'queen', 'man', 'woman', 'dog', 'cat', 'car', 'bus', 'apple', 'banana', 
         'computer', 'keyboard', 'mouse', 'monitor', 'london', 'paris', 'tokyo', 'rome']

np.random.seed(42)
# Simulate 50-dimensional GloVe-like vectors for our sample vocabulary
embeddings_dict = {word: np.random.randn(50) for word in words}

# Add some structure so similar words are actually closer
embeddings_dict['queen'] = embeddings_dict['king'] + np.random.randn(50) * 0.1
embeddings_dict['woman'] = embeddings_dict['man'] + np.random.randn(50) * 0.1
embeddings_dict['cat'] = embeddings_dict['dog'] + np.random.randn(50) * 0.1
embeddings_dict['paris'] = embeddings_dict['london'] + np.random.randn(50) * 0.1
embeddings_dict['banana'] = embeddings_dict['apple'] + np.random.randn(50) * 0.1

tokens = []
vectors = []

for word, vector in embeddings_dict.items():
    tokens.append(word)
    vectors.append(vector)

vectors = np.array(vectors)

print("Applying t-SNE dimensionality reduction...")
tsne = TSNE(n_components=2, random_state=42, perplexity=3)
vectors_tsne = tsne.fit_transform(vectors)

print("Plotting...")
plt.figure(figsize=(10, 8))
plt.scatter(vectors_tsne[:, 0], vectors_tsne[:, 1], c='blue', edgecolors='k')

for i, word in enumerate(tokens):
    plt.annotate(word, xy=(vectors_tsne[i, 0], vectors_tsne[i, 1]), xytext=(5, 2), 
                 textcoords='offset points', ha='right', va='bottom', fontsize=12)

plt.title('t-SNE Visualization of Word Embeddings')
plt.grid(True)
plt.show()
```
