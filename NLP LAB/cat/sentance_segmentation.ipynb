```python
import nltk
import re
from nltk.corpus import brown
```

```python
raw_text = brown.raw(categories='news')
```

```python
print("Sample Text:")
print(raw_text[:200])
```

```python
print("SENTENCE SEGMENTATION")

#simple regex pattern to segment setances by space, ?, !, and . characters.
regex_pattern = r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?|\!)\s'
regex_sentences = re.split(regex_pattern, raw_text)

nltk_sentences = nltk.sent_tokenize(raw_text)

print("\nComparing the performace:")
print(f"Regex sentences: {len(regex_sentences)}")
print(f"NLTK sentences: {len(nltk_sentences)}")
```

```python
print("WORD SEGMENTATION")

sample_sentence = nltk_sentences[0]

words = []
for token in sample_sentence.split():
    clean_word = token.strip(".,!?;:\"()[]")
    if clean_word:
        words.append(clean_word)

print(f"Words found: \n {words[:10]}... \n(Total: {len(words)})")
```

```python
print("PARAGRAPH SEGMENTATION")

#using regex to split paragraphs by two or more newlines
regex_paragraphs = re.split(r'\n{2,}', raw_text)

paragraphs = [p.strip() for p in regex_paragraphs if p.strip()]

print(f"Total Paragraphs Found: {len(paragraphs)}")
if paragraphs:
    print(f"First Paragraph: {paragraphs[0]}...")
```
