```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import os
```

```python
filename = r'C:\Users\vishn\Desktop\Programs\SEM 8\NLP LAB\document.txt'
```

```python
print("=== 1. Reading and Tokenizing ===")

all_tokens = []

with open(filename, 'r', encoding='utf-8-sig') as file:
    lines = file.readlines()
    for i, line in enumerate(lines):
        tokens = word_tokenize(line)
        all_tokens.extend(tokens)
        print(f"Line {i+1} tokens: {tokens}")
```

```python
print("\n=== 2. Stemming ===")
ps = PorterStemmer()

stemmed_words = [ps.stem(word) for word in all_tokens]

print("First 20 stemmed words:", stemmed_words[:20]) 

with open('stemmed_output.txt', 'w') as f:
    f.write(' '.join(stemmed_words))
```

```python
print("\n=== 3. Lemmatization ===")
lemmatizer = WordNetLemmatizer()

lemmatized_output = [lemmatizer.lemmatize(word) for word in stemmed_words]
print("First 20 lemmatized words:", lemmatized_output[:20])
```

```python
print("\n=== 4. Stop Words Removal ===")

stop_words = set(stopwords.words('english'))

filtered_text = [word for word in all_tokens if word.lower() not in stop_words]

clean_text_display = ' '.join(filtered_text)

print(f"Stop words count in NLTK list: {len(stop_words)}")
print(clean_text_display[:200])
```
