```python
import nltk
from nltk.sem.logic import LogicParser

lp = LogicParser()
```

```python
# a. If Angus sings, it is not the case that Bertie sulks.
expr_a = lp.parse('P -> -Q')

# b. Cyril runs and barks.
expr_b = lp.parse('R & S')

# c. It will snow if it doesn’t rain. (Note: "A if B" is B -> A)
expr_c = lp.parse('-A -> B')

# d. It’s not the case that Irene will be happy if Olive or Tofu comes.
expr_d = lp.parse('-((D | E) -> C)')

# e. Pat didn’t cough or sneeze. (De Morgan's Law applies here: neither A nor B)
expr_e = lp.parse('-(F | G)')

# f. If you don’t come if I call, I won’t come if you call.
# Structure: (If I call -> You don't come) -> (If You call -> I won't come)
expr_f = lp.parse('(I -> -H) -> (K -> -J)')

print("1a:", expr_a)
print("1b:", expr_b)
print("1c:", expr_c)
print("1d:", expr_d)
print("1e:", expr_e)
print("1f:", expr_f)
```

```python
# a. Angus likes Cyril and Irene hates Cyril.
fol_a = lp.parse('likes(angus, cyril) & hates(irene, cyril)')

# b. Tofu is taller than Bertie.
fol_b = lp.parse('taller(tofu, bertie)')

# c. Bruce loves himself and Pat does too.
fol_c = lp.parse('loves(bruce, bruce) & loves(pat, pat)')

# d. Cyril saw Bertie, but Angus didn’t.
fol_d = lp.parse('saw(cyril, bertie) & -saw(angus, bertie)')

# e. Cyril is a four-legged friend.
fol_e = lp.parse('fourlegged(cyril) & friend(cyril)')

# f. Tofu and Olive are near each other.
fol_f = lp.parse('near(tofu, olive) & near(olive, tofu)')

print("2a:", fol_a)
print("2b:", fol_b)
print("2c:", fol_c)
print("2d:", fol_d)
print("2e:", fol_e)
print("2f:", fol_f)
```

```python
# a. Angus likes someone and someone likes Julia.
quan_a = lp.parse('exists x. likes(angus, x) & exists y. likes(y, julia)')

# b. Angus loves a dog who loves him.
quan_b = lp.parse('exists x. (dog(x) & loves(angus, x) & loves(x, angus))')

# c. Nobody smiles at Pat.
quan_c = lp.parse('-exists x. smiles(x, pat)')

# d. Somebody coughs and sneezes.
quan_d = lp.parse('exists x. (coughs(x) & sneezes(x))')

# e. Nobody coughed or sneezed.
quan_e = lp.parse('-exists x. (coughed(x) | sneezed(x))')

# f. Bruce loves somebody other than Bruce.
quan_f = lp.parse('exists x. ((x != bruce) & loves(bruce, x))')

# g. Nobody other than Matthew loves Pat.
quan_g = lp.parse('all x. ((x != matthew) -> -loves(x, pat))')

# h. Cyril likes everyone except for Irene.
quan_h = lp.parse('all x. ((x != irene) -> likes(cyril, x))')

# i. Exactly one person is asleep.
# (Someone is asleep AND anyone who is asleep must be that person)
quan_i = lp.parse('exists x. (asleep(x) & all y. (asleep(y) -> (y = x)))')

print("3a:", quan_a)
print("3b:", quan_b)
print("3c:", quan_c)
print("3d:", quan_d)
print("3e:", quan_e)
print("3f:", quan_f)
print("3g:", quan_g)
print("3h:", quan_h)
print("3i:", quan_i)
```

```python
# a. feed Cyril and give a capuccino to Angus
lam_a = lp.parse(r'\x. (feed(x, cyril) & give(x, angus, capuccino))')

# b. be given "War and Peace" by Pat
# Passive construction: Pat gives x War_and_Peace
lam_b = lp.parse(r'\x. give(pat, x, war_and_peace)')

# c. be loved by everyone
# x is the entity being loved
lam_c = lp.parse(r'\x. all y. loves(y, x)')

# d. be loved or detested by everyone
lam_d = lp.parse(r'\x. all y. (loves(y, x) | detests(y, x))')

# e. be loved by everyone and detested by no-one
lam_e = lp.parse(r'\x. (all y. loves(y, x) & -exists z. detests(z, x))')

print("4a:", lam_a)
print("4b:", lam_b)
print("4c:", lam_c)
print("4d:", lam_d)
print("4e:", lam_e)
```
