```python
import re
import random

def respond(user_input, patterns):
    """
    Scans the user_input against the list of patterns.
    If a match is found, returns the corresponding response with substitutions.
    """
    for pattern, response in patterns:
        match = re.match(pattern, user_input, re.IGNORECASE)
        if match:
            # If the response contains groups like {0}, replace them with captured text
            # We use match.groups() to find the captured segments (e.g., 'sad', 'depressed')
            if '{0}' in response:
                # Get the specific captured group (e.g., 'sad')
                captured_word = match.group(1)
                return response.format(captured_word)
            else:
                return response
    
    # Default fallback if no pattern matches
    return "Tell me more about that."
```

```python
tech_patterns = [
    # Matches: "I found a bug", "I found a vulnerability"
    (r".*found a (bug|vulnerability).*", "GREAT WORK! WHAT IS THE SEVERITY OF THE {0}?"),
    
    # Matches: "The server is down", "The server is slow"
    (r".*server is (down|slow|crashed).*", "HAVE YOU CHECKED THE LOGS FOR THE {0} SERVER?"),
    
    # Matches: "I used Python", "I used Java"
    (r".*used (Python|Java|C\+\+).*", "INTERESTING CHOICE. WHY DID YOU CHOOSE {0} FOR THIS TASK?"),
    
    # Matches: "It is not working"
    (r".*not working.*", "WHAT IS THE ERROR MESSAGE YOU ARE GETTING?"),
    
    (r"quit|exit|bye", "HAPPY CODING!"),
]

print("Tech Support rules loaded.")
```

```python
def start_chat(patterns):
    print("--- Eliza Chatbot Initialized ---")
    print("(Type 'quit' to exit)\n")
    
    while True:
        # Get user input
        user_input = input("You: ")
        
        # Check for exit condition
        if user_input.lower() in ['quit', 'exit', 'bye']:
            print("Bot: GOODBYE!")
            break
            
        # Get response
        response = respond(user_input, patterns)
        print(f"Bot: {response}")

# Select which domain you want to run:
# Change this to 'tech_patterns' to run the Tech bot
selected_patterns = tech_patterns 

start_chat(selected_patterns)
```
