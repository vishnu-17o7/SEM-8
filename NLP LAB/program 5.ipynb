```python
# Install and import NLTK
# Uncomment the next line if NLTK is not installed
# !pip install nltk

import nltk
from nltk.corpus import brown
from collections import Counter

# Download Brown corpus if not available
nltk.download('brown')
```

## Observations on Brown Corpus
- The Brown Corpus is a large collection of texts from various genres, compiled in 1961.
- It contains about 1 million words, divided into 500 samples of about 2000 words each.
- It has 15 categories, representing different text types.
- Useful for linguistic studies, stylistics, and NLP tasks.


```python
# List all categories (sections)
categories = brown.categories()
print("Brown Corpus Categories/Genres:")
for i, cat in enumerate(categories, 1):
    print(f"{i}. {cat}")
```

```python
# Categorize documents: List fileids for each category
print("Documents (fileids) per Category:")
for cat in categories:
    files = brown.fileids(categories=[cat])
    print(f"{cat}: {[f[:10] + '...' if len(f) > 10 else f for f in files[:5]]}")  # Show first 5
    print(f"   Total files: {len(files)}\n")
```

```python
# Choose a section: 'news'
chosen_category = 'news'
print(f"Chosen Category: {chosen_category}\n")

# Number of words
words = brown.words(categories=[chosen_category])
num_words = len(words)
print(f"Number of words in '{chosen_category}': {num_words}")

# Number of sentences
sents = brown.sents(categories=[chosen_category])
num_sents = len(sents)
print(f"Number of sentences in '{chosen_category}': {num_sents}")

# Categories (it's just the one we chose)
print(f"Categories: {[chosen_category]}")

# Genres (same as categories)
print(f"Genres: {[chosen_category]}")

# Modals: Count modal verbs
modals = ['can', 'could', 'may', 'might', 'must', 'shall', 'should', 'will', 'would']
modal_counter = Counter()
for word in words:
    if word.lower() in modals:
        modal_counter[word.lower()] += 1
total_modals = sum(modal_counter.values())
print(f"Total modals: {total_modals}")
print("Modal counts:")
for modal, count in modal_counter.items():
    print(f"  {modal}: {count}")
```

```python
# Count 'wh' words in the chosen category
wh_words = ['what', 'when', 'where', 'why', 'who', 'whom', 'whose', 'which']
wh_counter = Counter()
for word in [w.lower() for w in words]:
    if word in wh_words:
        wh_counter[word] += 1
total_wh = sum(wh_counter.values())
print(f"Total 'wh' words in '{chosen_category}': {total_wh}")
print("'Wh' word counts:")
for wh, count in wh_counter.items():
    print(f"  {wh}: {count}")
```

```python
# Explore corpus functionalities

# fileids()
print("fileids(): ", brown.fileids()[:5])  # First 5

# fileids([categories])
print(f"fileids(['{chosen_category}']): ", brown.fileids([chosen_category])[:3])

# categories()
print("categories(): ", brown.categories())

# categories([fileids])
example_file = brown.fileids([chosen_category])[0]
print(f"categories(['{example_file}']): ", brown.categories([example_file]))

# raw()
print("len(raw()): ", len(brown.raw()))

# raw(fileids=[f1,f2,f3])
examples = brown.fileids()[:2]
print(f"len(raw(fileids={examples})): ", len(brown.raw(fileids=examples)))

# raw(categories=[c1,c2])
print(f"len(raw(categories=['{chosen_category}'])): ", len(brown.raw(categories=[chosen_category])))

# words()
print("len(words()): ", len(brown.words()))

# words(fileids=[f1,f2,f3])
print(f"len(words(fileids={examples[:1]})): ", len(brown.words(fileids=examples[:1])))

# words(categories=[c1,c2])
print(f"len(words(categories=['{chosen_category}'])): ", len(brown.words(categories=[chosen_category])))

# sents()
print("len(sents()): ", len(brown.sents()))

# sents(fileids=[f1,f2,f3])
print(f"len(sents(fileids={examples[:1]})): ", len(brown.sents(fileids=examples[:1])))

# sents(categories=[c1,c2])
print(f"len(sents(categories=['{chosen_category}'])): ", len(brown.sents(categories=[chosen_category])))

# abspath(fileid)
print("abspath('ca01'):", brown.abspath('ca01'))

# encoding(fileid)
print("encoding('ca01'):", brown.encoding('ca01'))

# open(fileid)
with brown.open('ca01') as f:
    print("open('ca01'): First 100 chars -", repr(f.read(100)))

# root()
print("root(): ", brown.root)

# readme()
readme = brown.readme()
print("readme() length: ", len(readme))
print("First 200 chars of readme:", repr(readme[:200]))
```
