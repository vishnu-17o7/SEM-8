```python
# Import necessary libraries
import nltk
from nltk import word_tokenize, sent_tokenize, FreqDist
from nltk.draw import dispersion_plot
import matplotlib.pyplot as plt

# Download required NLTK data
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# Download gutenberg corpus if needed
nltk.download('gutenberg')

# Load the text from NLTK Gutenberg corpus
from nltk.corpus import gutenberg
text = gutenberg.raw('blake-poems.txt')

print("Corpus loaded successfully.")
print(f"Text length: {len(text)} characters")
print(f"Number of words in corpus: {len(gutenberg.words('blake-poems.txt'))}")
```

```python
# Tokenize the text into words
tokens = word_tokenize(text)
total_tokens = len(tokens)

print(f"Total number of tokens: {total_tokens}")
print("First 20 tokens:", tokens[:20])
```

```python
# Tokenize into sentences
sentences = sent_tokenize(text)
total_sentences = len(sentences)

print(f"Total number of sentences: {total_sentences}")
print("First sentence:", sentences[0])
```

```python
# Create frequency distribution
freq_dist = FreqDist(tokens)

# (i) Display frequency of all words
print("Frequency distribution of all words:")
for word, freq in freq_dist.items():
    print(f"{word}: {freq}")

print("\nTotal unique words:", len(freq_dist))
```

```python
# (ii) Find and display frequency of a specific word
specific_word = "the"  # You can change this word
if specific_word in freq_dist:
    print(f"Frequency of '{specific_word}': {freq_dist[specific_word]}")
else:
    print(f"'{specific_word}' not found in the text.")
```

```python
# Find occurrence of a particular word
particular_word = "love"  # Change this to any word you want to check
occurrences = tokens.count(particular_word)
print(f"Occurrences of '{particular_word}': {occurrences}")
```

```python
# Get top 5 most common words
top_5 = freq_dist.most_common(5)

print("Top 5 words with highest frequency:")
for word, freq in top_5:
    print(f"{word}: {freq}")
```

```python
# Create dispersion plot for the top 5 words
words_to_plot = [word for word, freq in top_5]

print("Creating dispersion plot for top 5 words...")
dispersion_plot(tokens, words_to_plot)
plt.show()
```
