```python
# Import necessary libraries
import nltk
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import gutenberg

# Download required NLTK data
nltk.download('punkt')
nltk.download('gutenberg')

# Load sample text from document.txt
with open('document.txt', 'r', encoding='utf-8-sig') as file:
    sample_text = file.read()

print("Sample text loaded successfully.")
print(f"Text length: {len(sample_text)} characters")
```

```python
# (1) Hyphen ranges [2-5] and [b-f]
print("=== (1) Hyphen ranges ===")

# Find digits in range 2-5
digits_2_5 = re.findall(r'[2-5]', sample_text)
print(f"Digits 2-5 found: {digits_2_5}")

# Find letters in range b-f
letters_b_f = re.findall(r'[b-f]', sample_text.lower())
print(f"Letters b-f found: {letters_b_f}")

# Example with custom text
custom_text = "The quick brown fox jumps over 12345 lazy dogs."
digits_range = re.findall(r'[2-5]', custom_text)
letters_range = re.findall(r'[b-f]', custom_text.lower())
print(f"\nCustom text: {custom_text}")
print(f"Digits 2-5: {digits_range}")
print(f"Letters b-f: {letters_range}")
```

```python
# (2) ^ caret symbol
print("\n=== (2) ^ caret symbol ===")

# Find words that start lines (using multiline)
lines = sample_text.split('\n')
print("First few lines of text:")
for i, line in enumerate(lines[:5]):
    if line.strip():
        print(f"Line {i+1}: {line.strip()}")

# Find words starting with 'L' at beginning of lines
l_start_lines = re.findall(r'^L\w*', sample_text, re.MULTILINE)
print(f"\nWords starting with 'L' at line beginnings: {l_start_lines}")

# Find sentences starting with 'Large'
large_sentences = re.findall(r'^Large.*?(?:\.|$)', sample_text, re.MULTILINE)
print(f"Sentences starting with 'Large': {large_sentences[:2]}")  # Show first 2
```

```python
# (3) ? operator for optional preceding elements
print("\n=== (3) ? operator ===")

# Find 'neural' with optional 'deep ' before it
neural_pattern = re.findall(r'(?:deep )?neural', sample_text, re.IGNORECASE)
print(f"Patterns with optional 'deep ' before 'neural': {neural_pattern}")

# Find 'un' or 'in' followed by optional 'fair' or 'structured'
un_in_patterns = re.findall(r'\b(un|in)(?:fair|structured)?\b', sample_text, re.IGNORECASE)
print(f"'un'/'in' patterns with optional words: {un_in_patterns}")

# Example with colors: optional 'light' before color names
color_text = "I like red, light blue, green, and light yellow."
colors = re.findall(r'(?:light )?(red|blue|green|yellow)', color_text, re.IGNORECASE)
print(f"\nColor text: {color_text}")
print(f"Colors with optional 'light': {colors}")
```

```python
# (4) Kleene * operator
print("\n=== (4) Kleene * operator ===")

# Sheep language example as specified
sheep_text = """consider the language of certain sheep, which consists of strings that look like the
following:
baa!
baaa!
baaaa!
baaaaa!"""

print("Sheep language text:")
print(sheep_text)

# Match sheep sounds with zero or more 'a's
sheep_pattern = re.findall(r'ba*!', sheep_text)
print(f"\nSheep sounds matched: {sheep_pattern}")

# In the main text, find words with optional repeated letters
repeated_letters = re.findall(r'\b\w*(\w)\1*\w*\b', sample_text)
print(f"\nWords with repeated letters in sample text: {repeated_letters[:10]}")  # First 10
```

```python
# (5) Kleene + operator
print("\n=== (5) Kleene + operator ===")

# Find words with one or more vowels
vowel_words = re.findall(r'\b[\w]*[aeiouAEIOU]+[\w]*\b', sample_text)
print(f"Words with one or more vowels: {vowel_words[:15]}")  # First 15

# Find sequences of one or more digits
number_sequences = re.findall(r'\d+', sample_text)
print(f"\nNumber sequences: {number_sequences}")

# Find words with one or more 'l's
l_words = re.findall(r'\b\w*l+\w*\b', sample_text, re.IGNORECASE)
print(f"\nWords with one or more 'l's: {l_words[:10]}")  # First 10
```

```python
# (6) . (Dot) operator
print("\n=== (6) . (Dot) operator ===")

# Find three-letter words with any middle character
three_letter = re.findall(r'\b\w.\w\b', sample_text)
print(f"Three-letter words: {three_letter[:10]}")  # First 10

# Find patterns like 't.e' (t followed by any char followed by e)
t_e_patterns = re.findall(r'\bt.e\b', sample_text, re.IGNORECASE)
print(f"\nPatterns t.e: {t_e_patterns}")

# Find words containing 'a.b' pattern
a_b_patterns = re.findall(r'\b\w*a.b\w*\b', sample_text, re.IGNORECASE)
print(f"\nWords with 'a.b' pattern: {a_b_patterns[:5]}")  # First 5
```

```python
# (7) | pipe symbol
print("\n=== (7) | pipe symbol ===")

# Find 'machine' or 'deep' or 'neural'
ml_terms = re.findall(r'\b(machine|deep|neural)\b', sample_text, re.IGNORECASE)
print(f"ML terms (machine|deep|neural): {ml_terms}")

# Find 'and' or 'or' or 'but'
conjunctions = re.findall(r'\b(and|or|but)\b', sample_text, re.IGNORECASE)
print(f"\nConjunctions (and|or|but): {conjunctions[:15]}")  # First 15

# Find colors: red, blue, green, or yellow
colors_in_text = re.findall(r'\b(red|blue|green|yellow)\b', sample_text, re.IGNORECASE)
print(f"\nColors found: {colors_in_text}")
```

```python
# (8) Word boundary \b
print("\n=== (8) Word boundary \b ===")

# Find whole words 'the'
the_words = re.findall(r'\bthe\b', sample_text, re.IGNORECASE)
print(f"Occurrences of whole word 'the': {len(the_words)}")

# Find words starting with 'm'
m_words = re.findall(r'\bm\w*\b', sample_text, re.IGNORECASE)
print(f"\nWords starting with 'm': {m_words[:10]}")  # First 10

# Find words ending with 's'
s_words = re.findall(r'\b\w*s\b', sample_text, re.IGNORECASE)
print(f"\nWords ending with 's': {s_words[:10]}")  # First 10
```

```python
# (9) Non-word boundary \B
print("\n=== (9) Non-word boundary \B ===")

# Find 'the' not at word boundaries (part of other words)
the_inside = re.findall(r'\Bthe\B', sample_text, re.IGNORECASE)
print(f"'the' inside words: {the_inside}")

# Find patterns where 'ing' appears inside words
ing_inside = re.findall(r'\Bing\b', sample_text, re.IGNORECASE)
print(f"\n'ing' at end of words: {ing_inside[:10]}")  # First 10

# Find 'un' at the start of words (word boundary before, non-boundary after)
un_prefix = re.findall(r'\bun\B', sample_text, re.IGNORECASE)
print(f"\n'un' as prefix: {un_prefix}")
```

```python
# (10) Explanation of /[ˆa-zA-Z][tT]he[ˆa-zA-Z]/
print("\n=== (10) Pattern /[ˆa-zA-Z][tT]he[ˆa-zA-Z]/ ===")

# The pattern: non-letter, then 'the' (case insensitive), then non-letter
# Note: ˆ inside [] means negation, so [ˆa-zA-Z] means any character except letters
pattern_10 = r'[ˆa-zA-Z][tT]he[ˆa-zA-Z]'

# Create test text to demonstrate
test_text = "The cat is on the mat. They're going to the store. Thelma is here."
matches = re.findall(pattern_10, test_text)
print(f"Test text: {test_text}")
print(f"Matches for pattern: {matches}")
print("\nExplanation: This finds 'the' surrounded by non-letters (punctuation, spaces, etc.)")
print("It matches ' the ' and ' the.' but not 'the' inside 'They're' or 'Thelma'")

# Apply to sample text
sample_matches = re.findall(pattern_10, sample_text)
print(f"\nMatches in sample text: {len(sample_matches)} occurrences")
print(f"First few matches: {sample_matches[:5]}")
```
