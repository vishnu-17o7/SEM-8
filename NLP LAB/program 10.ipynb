```python
import random
from collections import defaultdict
```

```python
vocab = ["i", "want", "to", "eat", "chinese", "food", "lunch", "spend"]
V = 1446   # vocabulary size from image
```

```python
bigram_counts = {
    "i":       {"i":6, "want":828, "to":1, "eat":10, "chinese":1, "food":1, "lunch":1, "spend":3},
    "want":    {"i":3, "want":1, "to":609, "eat":2, "chinese":7, "food":7, "lunch":6, "spend":2},
    "to":      {"i":3, "want":1, "to":5, "eat":687, "chinese":3, "food":1, "lunch":7, "spend":212},
    "eat":     {"i":1, "want":1, "to":3, "eat":1, "chinese":17, "food":3, "lunch":43, "spend":1},
    "chinese": {"i":2, "want":1, "to":1, "eat":1, "chinese":1, "food":83, "lunch":2, "spend":1},
    "food":    {"i":16, "want":1, "to":16, "eat":1, "chinese":2, "food":5, "lunch":1, "spend":1},
    "lunch":   {"i":3, "want":1, "to":1, "eat":1, "chinese":1, "food":2, "lunch":1, "spend":1},
    "spend":   {"i":2, "want":1, "to":2, "eat":1, "chinese":1, "food":1, "lunch":1, "spend":1}
}
```

```python
unigram_counts = {}
for w1 in bigram_counts:
    unigram_counts[w1] = sum(bigram_counts[w1].values())
```

```python
def add_one_bigram_prob(w1, w2):
    return bigram_counts[w1][w2] / (unigram_counts[w1] + V)
```

```python
w1 = random.choice(vocab)
w2 = random.choice(vocab)

print("Random Bigram:", (w1, w2))
```

```python
prob = add_one_bigram_prob(w1, w2)
print(f"P({w2} | {w1}) with Add-One Smoothing =", prob)
```

```python
continuation_counts = defaultdict(int)

for w1 in vocab:
    for w2 in vocab:
        if bigram_counts[w1][w2] > 1:
            continuation_counts[w2] += 1

total_continuations = sum(continuation_counts.values())
```

```python
def kneser_ney_bigram_prob(w1, w2, d=0.75):
    c_bigram = bigram_counts[w1][w2]
    c_unigram = unigram_counts[w1]

    lambda_w1 = (d * len(bigram_counts[w1])) / c_unigram
    p_cont = continuation_counts[w2] / total_continuations if total_continuations > 0 else 0

    return max(c_bigram - d, 0) / c_unigram + lambda_w1 * p_cont
```

```python
kn_prob = kneser_ney_bigram_prob(w1, w2)
print(f"P({w2} | {w1}) with Kneser-Ney Smoothing =", kn_prob)
```
