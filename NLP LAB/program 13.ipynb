```python
import langid

samples = {
    "Japanese": "これは日本語の文章です。",
    "German": "Dies ist ein deutscher Satz.",
    "Dutch": "Dit is een Nederlandse zin.",
    "English": "This is an English sentence.",
    "Swedish": "Detta är en svensk mening."
}

print("Language Detection using LangID\n")

for name, text in samples.items():
    detected_lang, confidence = langid.classify(text)
    print(f"Input Language: {name}")
    print(f"Detected Language Code: {detected_lang}")
    print(f"Confidence: {confidence}\n")
```
