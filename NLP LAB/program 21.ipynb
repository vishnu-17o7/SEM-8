# Program 21
1. Perform Named Entity Recognition (NER) using the spacy library.
2. Perform Part-of-Speech (POS) tagging using the NLTK library on a given sentence based on the Penn Treebank tagset.
3. Implement a Hidden Markov Model (HMM) POS tagger using NLTK and train it on the Brown corpus.

```python
import nltk
# Ensure required NLTK resources are available
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('brown')
nltk.download('universal_tagset')

import spacy
from nltk.corpus import brown
from nltk.tag import hmm

# ==========================================
# 1. NER using spacy
# ==========================================
print("1. Named Entity Recognition (NER) using SpaCy")
try:
    nlp = spacy.load("en_core_web_sm")
except OSError:
    print("Downloading en_core_web_sm for spacy...")
    import spacy.cli
    spacy.cli.download("en_core_web_sm")
    nlp = spacy.load("en_core_web_sm")

text = "Apple is looking at buying U.K. startup for $1 billion"
doc = nlp(text)
print(f"Text: '{text}'")
for ent in doc.ents:
    print(f" - Entity: {ent.text}, Label: {ent.label_}")

# ==========================================
# 2. POS Tagging using NLTK (Penn Treebank)
# ==========================================
print("\n2. Part-of-Speech (POS) tagging using NLTK")
sentence = "The quick brown fox jumps over the lazy dog."
tokens = nltk.word_tokenize(sentence)
tags = nltk.pos_tag(tokens)
print(f"Sentence: '{sentence}'")
print(tags)

# ==========================================
# 3. Implement HMM POS tagger using NLTK (Brown Corpus)
# ==========================================
print("\n3. HMM POS Tagger using NLTK on Brown Corpus")
# Load POS-tagged sentences from the Brown corpus (using universal tagset to simplify)
brown_tagged_sents = brown.tagged_sents(tagset='universal')

# Split data into training and testing sets
size = int(len(brown_tagged_sents) * 0.9)
train_sents = brown_tagged_sents[:size]
test_sents = brown_tagged_sents[size:]

print(f"Training set size: {len(train_sents)}")
print(f"Testing set size: {len(test_sents)}")

# Train an HMM-based POS tagger
print("Training HMM Tagger...")
trainer = hmm.HiddenMarkovModelTrainer()
hmm_tagger = trainer.train_supervised(train_sents)

# Evaluate
accuracy = hmm_tagger.evaluate(test_sents)
print(f"Tagger Accuracy on Test Set: {accuracy * 100:.2f}%")

# Predict POS tags for a sample sentence
sample_phrase = "The model is predicting tags well"
sample_tokens = nltk.word_tokenize(sample_phrase)
predicted_tags = hmm_tagger.tag(sample_tokens)
print(f"\nSample Phrase: '{sample_phrase}'")
print("Predicted Tags:", predicted_tags)
```
